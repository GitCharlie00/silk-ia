    <h3 style="font-weight:700;"> 3. Grafici sul dataset </h3>
    <h3> 3.1 Bilanciamento dei valori target </h3>
</div>

```python

data_class_0_len = len(data[data.classificazione == 0])
data_class_1_len = len(data[data.classificazione == 1])

fig, ax = plt.subplots(figsize=(12,8))
plt.title("Bilanciamento dei valori target", fontsize=20, fontweight='bold')
ax.bar(0,data_class_0_len, label='Non mangia', align="center")
ax.bar(1,data_class_1_len, label='Mangia', align="center")

ax.set(xticks=range(2), xlim=[-1, 2])
ax.set(ylim=[0, 1000])
ax.text(0-0.093,data_class_0_len + 10,str(data_class_0_len), color='blue', fontweight='bold', fontsize=20)
ax.text(1-0.093,data_class_1_len + 10,str(data_class_1_len), color='orange', fontweight='bold', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

ax.legend(fontsize=15)
ax.set_xlabel('Classi', fontsize=15)
ax.set_ylabel('Numero di occorrenze', fontsize=15)
plt.show()
```

<div style="color: black; background-color: white; width: 93%; padding: 2em 2em 0.2em 2em; font-family: Gill Sans, sans-serif;">
    <h3> 3.2 Distribuzione dei valori delle features </h3>
</div>

```python
fig, axs = plt.subplots(2,2, figsize=(22,16))
plt.suptitle("Distribuzione dei valori delle features", fontsize=23, fontweight='bold')

sns.boxenplot(ax=axs[0][0],y=data['presenza_foglie'],palette=["#4287f5"],width=0.6)
axs[0][0].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[0][0].tick_params(axis='y', labelsize=18)
axs[0][0].yaxis.label.set_size(18)

sns.boxenplot(ax=axs[0][1],y=data['assenza_foglie'],palette=["#e547ed"],width=0.6)
axs[0][1].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[0][1].tick_params(axis='y', labelsize=18)
axs[0][1].yaxis.label.set_size(18)

sns.boxenplot(ax=axs[1][0],y=data['presenza_bachi_sfondo'],palette=["#54ff8a"],width=0.6)
axs[1][0].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[1][0].tick_params(axis='y', labelsize=18)
axs[1][0].yaxis.label.set_size(18)

sns.boxenplot(ax=axs[1][1],y=data['assenza_bachi_sfondo'],palette=["#ffaa54"],width=0.6)
axs[1][1].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[1][1].tick_params(axis='y', labelsize=18)
axs[1][1].yaxis.label.set_size(18)
```

<div style="color: black; background-color: white; width: 93%; padding: 2em 2em 0.2em 2em; font-family: Gill Sans, sans-serif;">
    <h3> 3.3 Correlation Matrix </h3>
</div>

```python
plt.subplots(figsize=(12,8))

plt.title("Correlation Matrix", fontsize=20, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

corrMatrix = data.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

<div style="color: black; background-color: white; width: 93%; padding: 2em 2em 0.2em 2em; font-family: Gill Sans, sans-serif;">
    <h3> 3.4 Scatter Plot: presenza foglie, bachi e sfondo </h3>
</div>

```python
data_class_0 = data[data.classificazione == 0]
data_class_1 = data[data.classificazione == 1]

fig, ax = plt.subplots(figsize=(12,8))
plt.title("(Presenza bachi, presenza foglie e sfondo) -> Classificazione", fontsize=20, fontweight='bold')
ax.scatter(data_class_0['presenza_foglie'], data_class_0['presenza_bachi_sfondo'], s=50, c='b', marker='o', label='Non mangia')
ax.scatter(data_class_1['presenza_foglie'], data_class_1['presenza_bachi_sfondo'], s=50, c='r', marker='x', label='Mangia')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

ax.legend(fontsize=15)
ax.set_xlabel('Presenza foglie (%)', fontsize=15)
ax.set_ylabel('Presenza bachi e sfondo (%)', fontsize=15)
plt.show()
```

<div style="color: black; background-color: white; width: 93%; padding: 2em 2em 0.2em 2em; font-family: Gill Sans, sans-serif;">
    <h3> 3.5 Distribuzione delle features in base al target </h3>
</div>

```python
fig, axs = plt.subplots(2,2, figsize=(22,16))
plt.suptitle("Distribuzione delle features in base al target", fontsize=23, fontweight='bold')

sns.kdeplot(ax=axs[0][0], data=data, x='presenza_foglie',hue="classificazione", fill=True,palette=["#8000ff","#da8829"], alpha=.5, linewidth=0)
axs[0][0].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[0][0].set_xlabel("presenza_foglie")
axs[0][0].set_ylabel("")

sns.kdeplot(ax=axs[0][1], data=data, x='assenza_foglie',hue="classificazione", fill=True,palette=["#8000ff","#da8829"], alpha=.5, linewidth=0)
axs[0][1].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[0][1].set_xlabel("assenza_foglie")
axs[0][1].set_ylabel("")

sns.kdeplot(ax=axs[1][0], data=data, x='presenza_bachi_sfondo',hue="classificazione", fill=True,palette=["#8000ff","#da8829"], alpha=.5, linewidth=0)
axs[1][0].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[1][0].set_xlabel("presenza_bachi_sfondo")
axs[1][0].set_ylabel("")

sns.kdeplot(ax=axs[1][1], data=data, x='assenza_bachi_sfondo',hue="classificazione", fill=True,palette=["#8000ff","#da8829"], alpha=.5, linewidth=0)
axs[1][1].grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
axs[1][1].set_xlabel("assenza_bachi_sfondo")
axs[1][1].set_ylabel("")
```
